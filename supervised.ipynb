```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x = np.array([-3,-2,-1,0,1,2,3,4,5])
x = x + np.random.normal(0,0.1,len(x))

def target(x):
    a = 2
    b = -1
    return(a*x+b)

y = target(x) + np.random.normal(0,0.5,len(x))
```

```python
range_x = np.linspace(-10,10,10)
range_y_1 = target(range_x)
range_y_2 = 1.5*range_x + 1

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(13,10))
for i,x_i in enumerate(x):
    ax2.plot([x_i, x_i], [y[i],target(x_i)], 'g')
    ax1.plot([x_i, x_i], [y[i],1.5*x_i +1], 'r')
ax1.scatter(x,y, label="Données d'entraînement")
ax2.scatter(x,y, label="Données d'entraînement")
ax1.plot(range_x, range_y_2, 'r', label="Modèle 1")
ax2.plot(range_x, range_y_1, 'g', label="Modèle 2")
for ax in [ax1, ax2]:
    ax.grid(True, which='both')
    ax.legend()
    ax.set_xlim([x.min()-1, x.max()+1])
    ax.set_ylim([y.min()-1, y.max()+1])
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')

plt.show()
```
