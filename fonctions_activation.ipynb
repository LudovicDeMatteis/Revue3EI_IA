```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 16}

matplotlib.rc('font', **font)

path_dossier = ""
```

```python
##########################################################
##            Fonction activation seuil                 ##
##########################################################

th = 0

x = np.linspace(-5,5,1000)
y = x>th

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)

ax.plot(x,y,linewidth=3)
ax.grid()
plt.title("Fonction d'activation seuil")
ax.set_xlabel("s")
ax.set_ylabel("g(s)")
ax.axhline(linewidth=1, color="k")
ax.axvline(linewidth=1, color="k")

plt.gca().set_position([0, 0, 1, 1])
#plt.savefig(path_dossier+"function_echelon_centre.svg", bbox_inches='tight')
```

```python
##########################################################
##           Fonction activation sigmoid                ##
##########################################################

alpha = 1
x = np.linspace(-5,5,1000)
y = 1/(1+np.exp(-alpha*x))
y2 = np.tanh(x)
y3 = [i if i>0 else 0 for i in x]

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)

ax.plot(x,y3,linewidth=3,label="ReLU")
ax.plot(x,y,linewidth=3,label="Sigmoid")
ax.plot(x,y2,linewidth=3,label="Tangente hyperbolique")

ax.grid()
#plt.title("Fonction d'activation sigmo√Øde")
ax.set_xlabel("s")
ax.set_ylabel("$\sigma$(s)")
ax.axhline(linewidth=1, color="k")
ax.axvline(linewidth=1, color="k")

plt.legend()
plt.gca().set_position([0, 0, 1, 1])
#plt.savefig(path_dossier+"function_sigmoid.svg", bbox_inches='tight')
```
