```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import matplotlib.cm as cm
from matplotlib.colors import LinearSegmentedColormap

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 16}

matplotlib.rc('font', **font)

path_dossier = ""

# get colormap
ncolors = 256
color_array = plt.get_cmap('bwr_r')(range(ncolors))

# change alpha values
color_array[:,-1] = np.linspace(0.5,0.5,ncolors)

# create a colormap object
map_object = LinearSegmentedColormap.from_list(name='my_cmap',colors=color_array)

# register this new colormap with matplotlib
plt.register_cmap(cmap=map_object)
```

```python
import numpy as np
def perceptron(x,w,th=0):
    value = np.sum(np.multiply(x,w),axis=1)
    return((value>th)*1)

weigth = np.array([2,2])
input = np.array([[1,1],[0,0],[0,1],[1,0]])
```

```python
##########################################################
##                   Fonction ET                        ##
##########################################################

weigth = np.array([2,2])
result = perceptron(input,weigth,3)

x_val = input[:,0]
y_val = input[:,1]
labels = [1,0,0,0]

cdict = {0: 'red', 1: 'blue'}

fig, ax = plt.subplots()
for g in np.unique(labels):
    ix = np.where(labels == g)
    ax.scatter(x_val[ix], y_val[ix], c = cdict[g], label = g, s = 100)
plt.axline((1, 0.5), slope=-1, color="black", linestyle="--")

ax.legend()
ax.grid()
plt.title("Fonction ET réalisée par le perceptron")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.axhline(linewidth=1, color="k")
ax.axvline(linewidth=1, color="k")

plt.gca().set_position([0, 0, 1, 1])
plt.savefig(path_dossier+"perceptron_and_graph.svg", bbox_inches='tight')

plt.show()
```

```python
###### ReLU activation function #########

def neurone(input, weight):
    sum = np.dot(input,weight)
    out_val = sum if sum>0 else 0
    return(out_val)

def reseau(input):
    input_list = [1, input]
    weigth = [[0.5,-1],[0.5,-1]]
    out_layer1 = neurone(input_list,weigth[0])
    input_list = [1, out_layer1]
    out_layer2 = neurone(input_list,weigth[1])
    return(out_layer2)

list_x = np.linspace(-2,2)
list_y = [reseau(i) for i in list_x]

fig, ax = plt.subplots()
ax.plot(list_x,list_y)
#plt.axline((1, 0.5), slope=-1, color="black", linestyle="--")

#ax.legend()
ax.grid()
plt.title("Sortie de deux neurones ReLU")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.axhline(linewidth=1, color="k")
ax.axvline(linewidth=1, color="k")

#plt.gca().set_position([0, 0, 1, 1])
plt.savefig(path_dossier+"relu_reseau_5_15_1.svg", bbox_inches='tight')

plt.show()
```
